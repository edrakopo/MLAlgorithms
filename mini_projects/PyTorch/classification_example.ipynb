```python
import sklearn.datasets as datasets
import matplotlib.pyplot as plt

#create dataset:
dat, lbs = datasets.make_circles(n_samples=1000, shuffle=True, noise=0.05, random_state=True, factor=0.7)

plt.scatter(dat[:,0], dat[:,1], c=lbs)
plt.show()
```

```python
#construct a subclass of nn.Module, a PyTorch module for building a neural network.
import torch.nn as nn
import torch.nn.functional as F

#Subclassing nn.Module
class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        self.fc1=nn.Linear(2,100)
        self.fc2=nn.Linear(100,2)

    def forward(self, x):
      
        x=F.relu(self.fc1(x))
        x=self.fc2(x)
        return F.softmax(x, dim=1)
    
net=Net().cuda()

net
```
